**BUSINESS SALES PERFORMANCE**

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error as mse
```

```python
#reading data
df=pd.read_csv("https://www.dropbox.com/s/argivhofekhaoau/BusinessSales.csv?dl=1")
```

```python
df.head()
```

```python
df.describe
```

```python
df.info()
```

```python
df = df.fillna(df.mean())
```

```python
df = pd.get_dummies(df)
```

```python
df
```

```python
df.columns
```

```python
df=df[['training level', 'work experience', 'salary',
       'division_computer hardware', 'division_computer software',
       'division_office supplies', 'division_peripherals', 'division_printers',
       'level of education_associate\'s degree',
       'level of education_bachelor\'s degree',
       'level of education_high school', 'level of education_master\'s degree',
       'level of education_some college','sales']]
```

```python
df.head()
```

```python
df.corr()
```

```python
x=df.iloc[:,0:-1].values
y=df.iloc[:,-1:].values
```

```python
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
print(x_train)
print(y_train)
```

```python
print(len(x_train),len(x_test))
```

```python
#linear Regression
lr_regressor=LinearRegression()
lr_regressor.fit(x_train,y_train)
y_pred_lr=lr_regressor.predict(x_test)
print(r2_score(y_test,y_pred_lr))
print(mse(y_test,y_pred_lr)**0.5)
```

```python
print(lr_regressor.coef_,lr_regressor.intercept_)
```

```python
df.columns
```

```python
#Random Forest Regression
rf_regressor=RandomForestRegressor()
rf_regressor.fit(x_train,y_train)
y_pred_rf=rf_regressor.predict(x_test)
print(r2_score(y_test,y_pred_rf))
print(mse(y_test,y_pred_rf)**0.5)
```
